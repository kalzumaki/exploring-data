## TREEMAP

```python
import plotly.express as px
import webbrowser
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('../data/GNI2014.csv')
df.head()

fig = px.treemap(data_frame = df,
                 path = ['continent', 'iso3'],
                 values='population',
                 color = 'GNI',
                 color_continuous_scale = 'Bluyl')
```

```python
plt.axis('off')
fig.update_layout(margin_t=50, margin_l=25, margin_r=25, margin_b=25, width=800, height=600, title_text='GNI 2014', title_font_size=20)

fig.write_html('../data/treemap.html')
webbrowser.open('../data/treemap.html')
```

```python
import matplotlib.image as mpimg
fig.write_image('../data/treemap.png')
plt.imshow(mpimg.imread('../data/treemap.png'))
plt.show()
```
