## BUBBLE CHART


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/crimeRatesByState2005.csv')
df.head()
```

```python

sns.set_theme(rc = {"figure.figsize": (7, 7)})
sns.scatterplot(data=df, x='murder', y='burglary',size='population', sizes=(20, 4000), hue='state', alpha=0.5, legend=False)

plt.xlim(0,12)

for i in range(0, df.shape[0]):
    plt.text(x=df.murder[i], y=df.burglary[i], s=df.state[i], horizontalalignment= 'center', size='small', color='dimgray')
    
plt.show()
```
