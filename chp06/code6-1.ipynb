```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/mtcars.csv')
df
```

```python
df.plot.scatter(x='wt', y='mpg')
plt.show()
```

```python
df.plot.scatter(x='hp', y='mpg', s=50, c='red', marker='o', grid=True)
plt.show()
```

```python
vars = ['mpg', 'disp', 'wt', 'drat']
pd.plotting.scatter_matrix(df[vars])
plt.show()
```
