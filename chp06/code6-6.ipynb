```python
import pandas as pd
import matplotlib.pyplot as plt

late = pd.Series([5,8,7,9,4,6,12,13,8,6,6,4], index=list(range(1,13)))
late.plot(title='Late student per month', xlabel='month',ylabel='frequency',linestyle='solid', marker='o', grid=True)
plt.show()
```
