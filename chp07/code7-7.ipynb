## Data sampling and combination


```python
import pandas as pd
import itertools
df = pd.read_csv('../data/iris.csv')

df20= df.sample(n=20, random_state=123)
df20
```

```python
stratified = df.groupby('Species').apply(lambda x: x.sample(frac=0.2, random_state=123), include_groups=False)
stratified
```

```python
species = df.Species.unique()
comb = list(itertools.combinations(species, 2))
comb
```
