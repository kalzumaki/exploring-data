```python
import pandas as pd

favorite = pd.Series(['Winter', 'Summer', 'Spring', 'Summer', 'Summer', 'Fall', 'Fall', 'Summer', 'Spring', 'Spring'])
favorite
```

```python
favorite.value_counts()
```

```python
favorite.value_counts() / favorite.size
```

```python
fd = favorite.value_counts()

type(fd)
```

```python
fd['Summer']
```

```python
fd.iloc[2]
```

```python
import matplotlib.pyplot as plt

fd.plot.bar(xlabel = 'Seasons', ylabel = 'Frequency',
            rot = 30, title='Favorite Seasons')
plt.show()
```

```python
#vertical bar plot
fd.plot.barh(xlabel = 'Seasons', ylabel = 'Frequency',
            rot = 30, title='Favorite Seasons')
plt.subplots_adjust(left=0.2)
plt.show()
```

```python
fd.plot.pie()
plt.show()
```

```python
fd.plot.pie(ylabel = '', autopct='%1.0f%%', title="Favorite Seasons")
plt.show()
```
