## BOX PLOT ON IRIS

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/iris.csv')

df.boxplot(column='Petal_Length', by='Species', grid=False)
plt.suptitle('')
plt.show()
```

# SEABORN

```python
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.boxplot(data=df, showfliers=False)
sns.stripplot(data=df, color='k', alpha=0.5)
plt.show()
```
