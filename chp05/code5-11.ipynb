## PANDAS GRAPH

```python
import pandas as pd
import matplotlib.pyplot as plt

favorite = pd.Series(['Winter', 'Summer', 'Spring', 'Summer', 'Summer', 'Fall', 'Fall', 'Summer', 'Spring', 'Spring'])
fd = favorite.value_counts()

fd.plot.bar(xlabel='Seasons', ylabel='Frequency', title='Favorite Seasons', color='c', rot=30, grid=True, figsize=(6,4))
plt.subplots_adjust(bottom=0.2)
plt.show()
```
