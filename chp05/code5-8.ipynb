```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../sample-data/sample_car.csv')
dist = df['dist']
dist
```

```python
dist.plot.hist()
plt.show()
```

```python
dist.plot.hist(bins=10)
plt.show()
```

```python
dist.value_counts(bins=6, sort=False)
dist.plot.hist(bins=6, title='Braking Distance', xlabel='Distance', ylabel='Frequency', color='m')
plt.show()
```
