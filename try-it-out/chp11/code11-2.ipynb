Using the titanic dataset from the seaborn library, perform a hypothesis test to determine whether surviving is related to which cabin (pclass) a passenger stayed in, at a significance level of 0.05

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')
df
```

```python
# cabin (pclass) and survival (alive) columns, create a cross tabulation and check the content
crosstab = pd.crosstab(df['pclass'], df['alive'])
crosstab
```

```python
# Calculate the survival rate by cabin 

survival_rate_by_cabin = df.groupby('pclass')['alive'].value_counts(normalize=True).unstack()['yes']
print(survival_rate_by_cabin)
```

```python
# Check the expected frequencies in the cross tabulation
expected_frequencies = stats.chi2_contingency(crosstab)[1]
expected_frequencies
```

```python
# perform chi-square test, check the p-value
stats.chi2_contingency(crosstab)
```
