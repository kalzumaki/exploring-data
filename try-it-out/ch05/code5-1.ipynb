```python
import pandas as pd
import matplotlib.pyplot as plt

hobby = pd.Series(['climbing','fishing','golf','swimming', 'climbing', 'climbing', 'fishing', 'swimming', 'climbing', 'fishing', 'swimming', 'golf'])

hobby
```

```python
# create a frequency distribution table by hobby
hobby.value_counts().plot(kind='bar', xlabel='Hobby', ylabel='Frequency', rot=30)
```

```python
fig, axes = plt.subplots(ncols=2)

hobby.value_counts().plot(kind='bar', ax=axes[0], rot=30)
hobby.value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%', ylabel='')

plt.show()
```
